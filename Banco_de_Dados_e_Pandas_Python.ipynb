#Agora vai 3 - Criando um sql

#Criar o banco de dados local

```python
import sqlite3 #para poder mexer com sql
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
dados = pd.read_csv(url)

dados.head()
```

```python
#conexão com a base de dados
conn = sqlite3.connect('dados_cliente.sqlite')

#converter para sql
dados.to_sql('dados_cliente', conn, if_exists='replace', index=False)
```

```python
#visualizar o sql
tabela_db = pd.read_sql('SELECT * FROM dados_cliente', conn)
tabela_db.head()
```

# Atualizar registro


```python
#Identificar a linha
conn.execute('UPDATE dados_cliente SET Rendimento_anual = 300000 WHERE ID_Cliente = 6840104;')

tabela_db = pd.read_sql('SELECT * FROM dados_cliente', conn)
tabela_db.query('ID_Cliente == 6840104')
```

#Excluir o registro

```python
conn.execute('DELETE FROM dados_cliente WHERE ID_Cliente = 6850985;')

tabela_db = pd.read_sql('SELECT * FROM dados_cliente', conn)
```

#Criar um novo registro

```python
conn.execute('INSERT INTO dados_cliente (ID_Cliente, Idade, Grau_escolaridade, Estado_civil,Tamanho_familia) VALUES (6850985,33,"Doutorado","Solteiro",1)')

tabela_db = pd.read_sql('SELECT * FROM dados_cliente', conn)
```
